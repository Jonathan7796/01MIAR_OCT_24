#### Descripción
El conjunto de datos "valencia_pollution_dataset.csv" (fichero disponible adjunto a la actividad) proviene de una descarga de datos del servicio web del ayuntamiento de la ciudad de Valencia de datos abiertos.

Los datos recabados de este servicio web son de mediciones cada hora de una serie de estaciones de contaminación atmosférica.

https://valencia.opendatasoft.com/explore/dataset/estacions-contaminacio-atmosferiques-estaciones-contaminacion-atmosfericas/information/

Cada estación realiza una serie de mediciones acerca de la calidad del aire, y de ciertas sustancias que se consideran importantes para analizar la contaminación atmosférica en la ciudad.

#### Columnas para trabajar:
- objectid: id de la estación
- nombre: nombre de la estación
- dirección: dirección de la estación
- tipozona: tipo de la zona, urbana o sub urbana
- tipoemision: tráfico o fondo
- so2: niveles de dióxido de azufre
- no2: niveles de dióxido de nitrogeno
- o3: niveles de ozono
- co: niveles de monóxido de carbono
- pm10: niveles de partículas en suspensión hasta 10um
- pm25: niveles de partículas en suspensión hasta 2,5um
- fecha_carga: timestamp de la actualización de los datos
- calidad_ambiental: estimación de la calidad del aire

```python
# Your imports here...
#Librerias a utilizar en todo el analisis
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 01
Carga de datos de csv (valencia_pollution_dataset.csv) en un DataFrame

## Nota: 
Para la actividad he generado una carpeta llamada "dataset_video" dentro del repositorio local 01MIAR_OCT_24, en esta he subido el dataset "valencia_pollution_dataset.csv" de esta forma a continuacion construyo la ruta para obtener acceso a los datos. 

```python
# Your solution here...
#Se construye la ruta del dataset
ruta_dataset = os.path.join("dataset_video", "valencia_pollution_dataset.csv")
#Se verifica si el dataset existe o no en la ruta
if os.path.exists(ruta_dataset):
    data_frame = pd.read_csv(ruta_dataset)
    print("Archivo cargado correctamente.")
    # Muestro las 3 primeras filas del DataFrame para saber si se cargo correctamente
    print(data_frame.head(3))   
else:
    print(f"El archivo no se encuentra en la ruta: {ruta_dataset}")
```

## 

## 02
- ¿Cuál es el rango temporal del dataset?

```python
# Your solution here...
#Rango Temporal
#Uso "to_datetime" para convertir la fecha en un tipo de dato manipulable
data_frame['fecha_carga'] = pd.to_datetime(data_frame['fecha_carga'],errors = 'coerce')
#Obtengo valores minimos y maximos para obtener el rango temporal
fecha_min = data_frame['fecha_carga'].min()
fecha_max = data_frame['fecha_carga'].max()
rango_temporal = fecha_max - fecha_min
print(f"El rango temporal obtenido del Dataset es: {rango_temporal}")
```

- Obtención del número de estaciones que tienen un sensor para medir monóxido de carbono

Para calcular el numero de estaciones que tienen un sensor para medir CO, se asume que las celdas donde se obtiene NaN en la columna de CO, son los que no tienen sensor de medicion.

```python
#Uso de notna() para filtrar filas donde 'co' no es nulo
filtrado_datos = data_frame[data_frame['co'].notna()]  
#Uso de nunique() para filtrar los id de cada estacion y obtener las estaciones sin repeticiones
filtrado_objectid = filtrado_datos['objectid'].nunique()
print(f"El numero de estaciones que disponen de un sensor para la mediciones de CO son: {filtrado_objectid}")
```

## 03

- Número de estación con la media más alta de mediciones de monóxido de carbono

```python
# Your solution here...
#Hago uso de 'groupby' para agrupar los datos por el identificador 'objectid' y calcular sus medias
#Esto lo realizo en el dataframe ya filtrado que no posee "objectid" con mediciones en NaN
media_por_estacion = filtrado_datos.groupby('objectid')['co'].mean()
# Obtener la estación con la media más alta
estacion_con_media_maxima = media_por_estacion.idxmax()
print(f"El 'objectid' de la estacion con la media mas alta de mediciones de CO es:{estacion_con_media_maxima} ")
```

## 04
Realizar el histograma de dióxido de nitrógeno de la estación del punto anterior

```python
# Your solution here...
#Filtro los datos de 'no2' para no tener valores nulos
filtrado_datos_no2 = data_frame[data_frame['no2'].notna()]
#Obtengo los valores de 'no2' de la estacion anterior
df_estacion_26 = filtrado_datos_no2[filtrado_datos_no2['objectid'] == estacion_con_media_maxima]
mediciones_no2 = df_estacion_26['no2'].dropna()

#Realizo el histograma mediante libreria 'matplotlib'
plt.figure(figsize=(10, 6))
plt.hist(mediciones_no2, bins=20, color='skyblue', edgecolor='black')
plt.title(f'Histograma de Mediciones de NO2 para la estacion: {estacion_con_media_maxima}')
plt.xlabel('Concentración de NO2')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

## 05
Realizar gráfica con  la línea temporal de las mediciones de ozono de la estación del punto anterior. Pista, la columna 'fecha_carga' habría que cambiarla de tipo.

```python
# Your solution here...
#Filtro los datos de 'o3' para no tener valores nulos
filtrado_datos_o3 = data_frame[data_frame['o3'].notna()]
#Convierto los datos de tiempo en celda 'fecha_carga' en datos manipulables
filtrado_datos_o3.loc[:, 'fecha_carga'] = pd.to_datetime(filtrado_datos_o3['fecha_carga'],errors = 'coerce')
#Obtengo los datos de mediciones de 'o3' en la estacion indicada
df_estacion_26 = filtrado_datos_o3[filtrado_datos_o3['objectid'] == estacion_con_media_maxima]
mediciones_o3 = df_estacion_26['o3'].dropna()

#Realizo la grafica haciendo uso de las mediciones_o3 en funcion de la fecha de carga
plt.figure(figsize=(10, 6))
plt.plot(df_estacion_26['fecha_carga'], mediciones_o3, color='green', marker='o', linestyle='-', markersize=4)
plt.title(f'Histograma de Mediciones de o3 para la estacion: {estacion_con_media_maxima}')
plt.xlabel('Fecha')
plt.ylabel('Concentración de Ozono')
plt.grid(True)
# Roto las fechas para mejor legibilidad
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```
