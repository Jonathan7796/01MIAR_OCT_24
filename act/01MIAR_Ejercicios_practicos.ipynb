## 02 - 1.5pt
- Escribe una función que reciba un DataFrame, le agregue una nueva columna llamada ‘Promedio’ que sea el promedio de sus columnas Luego, filtre las filas donde el valor de ‘Promedio’ sea mayor que 5 y lo ordene por la columna ‘Promedio’ en orden descendente

```python
# Your solution HERE !!!!
def func_mean(df): 
    # Agregar la columna 'Promedio' que es el promedio de las columnas numéricas
    df['Promedio'] = df.mean(axis=1)  # Promedio por fila (axis=1)
    # Filtrar las filas donde 'Promedio' sea mayor que 5
    df_filtrado = df[df['Promedio'] > 5]   
    # Ordenar el DataFrame por la columna 'Promedio' en orden descendente
    df_ordenado = df_filtrado.sort_values(by='Promedio', ascending=False).reset_index(drop = True)
    return df_ordenado
```

```python
# Test Use Case
source_dataframe = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 7, 6], 'C': [7, 12, 9]})
expected_dataframe = pd.DataFrame({'A': [2, 3], 'B': [7, 6], 'C': [12, 9], 'Promedio':[7.0, 6.0]})

output = func_mean(source_dataframe) # This calls your function named func_mean

display(expected_dataframe)
display(output)

assert(expected_dataframe.equals(output)) # This will fail if the dataframe is wrong
```

## 04 - 1.5pt
- Crear una función que reciba un array de NumPy y que reemplace todos los números primos por sus raices cuadradas
- Primo es aquel número natural mayor que 1, que solo es divisible por 1 y por si mismo

```python
# Your solution HERE !!!!

def es_primo(n):
    if n <= 1:
        return False
    for i in range(2, int(np.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

def prime_per_sqrt(array):
      # Iterar sobre todos los elementos del array, manteniendo la estructura original
    array = array.astype(float)
    for idx in np.ndindex(array.shape):  # np.ndindex recorre todos los índices del array
        if es_primo(array[idx]):
            array[idx] = np.round(np.sqrt(array[idx]),2)  # Reemplaza el número primo por su raíz cuadrada
    
    return array
```

```python
# Test Use Case
source_array = np.array([[20, 17, 33, 11], [3, 25, 5, 49], [18, 28, 32, 2], [9, 1, 23, 40]])
expected_output = np.array([[20, 4.12, 33, 3.31], [1.73, 25, 2.23, 49], [18, 28, 32, 1.41], [9, 1, 4.79, 40]])

output = prime_per_sqrt(source_array) # This calls your function named prime_per_sqrt

display(expected_output)
display(output)

assert(np.allclose(output, expected_output, atol = 0.01)) # This will fail if the array is wrong
```

El conjunto de datos "estaciones_bici.csv" (fichero disponible adjunto al notebook) proviene de una descarga de datos del servicio web de la empresa municipal Valenbici, dedicada al alquiler de bicicletas en Valencia.

Los datos recabados de este servicio web son de mediciones cada 10 minutos de las estaciones:

https://valencia.opendatasoft.com/explore/dataset/valenbisi-disponibilitat-valenbisi-dsiponibilidad/information/

Cada estación está compuesta por un número variable de bornetas (total) donde se pueden anclar las bicicletas. Los datos obtenidos de cada estación (registros) refleja el número de bornetas libres (free) y el número de bicicletas disponibles (available).

#### Columnas para trabajar:
- station: id de la estación
- total: número total de bornetas
- updated: timestamp del estado en de bicis en la estación
- available: número de bicis disponibles

#### Preguntas
1. Carga de datos de csv (estaciones_bici.csv) en un DataFrame, descripción de los datos
2. ¿Cuál es el rango temporal del dataset? Obtención del número de estaciones que tienen un total de 25 bornetas, y cuales son
3. Número de estación con la media más baja de bicis disponibles
4. Realizar el histograma de bicis disponibles de la estación del punto anterior
5. Realizar gráfica con  la línea temporal de bicis disponibles de la estación del punto anterior. Pista, es necesario cambiar el índice a uno tipo fecha

```python
# Your solution HERE !!!!
#Se construye la ruta del dataset
ruta_dataset = os.path.join("..","dataset_video", "estaciones_bici.csv")
#Se verifica si el dataset existe o no en la ruta
if os.path.exists(ruta_dataset):
    data_frame = pd.read_csv(ruta_dataset,sep=";")
    print("Archivo cargado correctamente.")
    # Muestro las 3 primeras filas del DataFrame para saber si se cargo correctamente
    print(data_frame.head(3))   
else:
    print(f"El archivo no se encuentra en la ruta: {ruta_dataset}")
```

```python
#Rango Temporal
#Uso "to_datetime" para convertir la fecha en un tipo de dato manipulable
data_frame['download_date'] = pd.to_datetime(data_frame['download_date'],errors = 'coerce')
#Obtengo valores minimos y maximos para obtener el rango temporal
fecha_min = data_frame['download_date'].min()
fecha_max = data_frame['download_date'].max()
rango_temporal = fecha_max - fecha_min
print(f"El rango temporal obtenido del Dataset es: {rango_temporal}")
```

```python
# Filtrar estaciones con 25 bornetas
estaciones_25_bornetas = data_frame[data_frame['total'] == 25]
# Número de estaciones con 25 bornetas
num_estaciones_25 = estaciones_25_bornetas.shape[0]
print(f"Número de estaciones con 25 bornetas: {num_estaciones_25}")
print("Estaciones con 25 bornetas:")
print(estaciones_25_bornetas[['station']])
```

```python
# Calcular la media de bicicletas disponibles por estación
media_bicis_disponibles = data_frame.groupby('station')['available'].mean()

# Encontrar la estación con la media más baja
estacion_media_baja = media_bicis_disponibles.idxmin()
media_baja = media_bicis_disponibles.min()

print(f"Estación con la media más baja de bicis disponibles: {estacion_media_baja}")
print(f"Media más baja de bicis disponibles: {media_baja}")
```

```python
# Filtrar datos de la estación con la media más baja
df_estacion_baja = data_frame[data_frame['station'] == estacion_media_baja]

# Crear el histograma de bicicletas disponibles
plt.figure(figsize=(10,6))
plt.hist(df_estacion_baja['available'], bins=20, color='blue', edgecolor='black')
plt.title(f"Histograma de Bicis Disponibles - Estación {estacion_media_baja}")
plt.xlabel("Bicis Disponibles")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

```python
# Convertir 'download_date' a índice de tipo fecha
df_estacion_baja = df_estacion_baja.set_index('download_date')

# Crear la gráfica con la línea temporal
plt.figure(figsize=(12,6))
plt.plot(df_estacion_baja.index, df_estacion_baja['available'], marker='o', color='green')
plt.title(f"Línea Temporal de Bicis Disponibles - Estación {estacion_media_baja}")
plt.xlabel("Fecha")
plt.ylabel("Bicis Disponibles")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
